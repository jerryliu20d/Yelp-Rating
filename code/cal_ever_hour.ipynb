```python
bus = pd.read_json(path_train_bus, lines=True, orient="columns")
```

```python
rbus = bus[['business_id','is_open','categories']]
```

```python
com = pd.concat([rbus,h,c],axis=1)
```

```python
print(type(com['weekday'].values[0]))
com.head()
```

```python
com['weekday_ever'] = com['weekday_hour']/com['weekday']
com['weekend_ever'] = com['weekend_hour']/com['weekend']
```

```python
com['weekend_ever'].values
com.loc[com['weekend']==0,'weekend_ever'] = 0
```

```python
com['weekday_ever'].values
com.loc[com['weekday']==0,'weekday_ever'] = 0
```

```python
com['week_ever'] = (5*com['weekday_ever']+2*com['weekend_ever'])/7
```
